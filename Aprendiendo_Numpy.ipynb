Se puede notar que el 5to valor aparece como nan, en el siguiente caso por ser un array de un tamaño pequeño es facil vizualizar, pero no sería viable hacer este mismo proceso para arrays mayores, o en multiples array. Para tenern una visión podemos evaluar cuantos nan  exiten en un arreglo.

Para saber cuantos nan hay en un arreglo se usa la función isnan de numpy

```python
np.isnan(Kaliningrad)
```

Usandolo de esta manera no sería tan útil

Entonces, sabiendo que los valores false son interpretados como 0 y los valores tipo true como 1, se puede usar la función SUM para saber el total de valores NAN.

```python
sum(np.isnan(Kaliningrad))
```

De esta manera podriamos evaluar todo un dataset para saber quantos valores nan exiten dentro del él.

Ahora que ya sabemos que solo tenemos un solo valor nan vamos tratar ese valor faltante. Una de las formas de completar ese valor faltante, puede ser interpolando con sus valores mas cercanos usando una media entre ellos: Pudiendo usar el valor antes y después del nan como referencia.

La primera forma sería, sabiendo que el nan se encuentra en el index 4, podemos usar el index 3 y 5 para sacar la media

```python
Kaliningrad[4] = (Kaliningrad[3]+Kaliningrad[5])/2
Kaliningrad[4]
```

En numpy hay una funcion llamada mean, que devuelve la media de de una lista de valores.


```python
Kaliningrad[4]= np.mean([Kaliningrad[3],Kaliningrad[5]])
Kaliningrad[4]
```
